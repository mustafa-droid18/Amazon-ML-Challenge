# Data Cleaning and Preprocessing

```python
df_train=df_train.fillna(' ')
```

```python
words = nltk.word_tokenize(df_train['TITLE'][1])
print (words)
```

```python
df= df_train.iloc[:5].copy()
df['TITLE']=df['TITLE'].apply(lambda x: nltk.word_tokenize(x) if isinstance(x, str) else x)
df['TITLE']=df['TITLE'].apply(lambda x: [word for word in x if word.isalnum()])
df['TITLE']=df['TITLE'].apply(lambda x: [word for word in x if word not in set(stopwords.words("english"))])
df
```

```python
columns = ['TITLE','BULLET_POINTS','DESCRIPTION']
for i in columns:
    df[i]=df[i].apply(lambda x: nltk.word_tokenize(x) if isinstance(x, str) else x)
    df[i]=df[i].apply(lambda x: [word for word in x if word.isalnum()])
    df[i]=df[i].apply(lambda x: [word for word in x if word not in set(stopwords.words("english"))])
df
```

```python
def displyafull(df, col):
    for i in range(5):
        print(i, '. ', df[col][i])

displyafull(df, 'TITLE')
print(' ')
displyafull(df, 'BULLET_POINTS')
print(' ')
displyafull(df, 'DESCRIPTION')
```

```python
displyafull(df_train, 'TITLE')
print(' ')
displyafull(df_train, 'BULLET_POINTS')
print(' ')
displyafull(df_train, 'DESCRIPTION')
```
