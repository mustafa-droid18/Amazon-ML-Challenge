```python
import pandas as pd
import numpy as np 
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps = PorterStemmer()
from nltk.stem import WordNetLemmatizer
lm= WordNetLemmatizer()
```

```python
df_train = pd.read_csv('dataset/train.csv')
df_train.head(10)
df_train=df_train.fillna(' ')
```

```python
def Batch(upper,lower=0):
    df=df_train.iloc[lower:upper].copy()
    columns = ['TITLE','BULLET_POINTS','DESCRIPTION']
    for i in columns:
        df[i]=df[i].apply(lambda x: nltk.word_tokenize(x) if isinstance(x, str) else x)
        df[i]=df[i].apply(lambda x: [word for word in x if word.isalnum()])
        df[i]=df[i].apply(lambda x: [word for word in x if word not in set(stopwords.words("english"))])
        #df[i]=df[i].apply(lambda x: [ps.stem(word) for word in x])
        df[i]=df[i].apply(lambda x: [lm.lemmatize(word) for word in x])
        df[i]=df[i].apply(lambda x: [word.lower() for word in x])
    df.to_csv('new_dataset/dataset_{}_{}.csv'.format(lower,upper), index=False) 
    return df
```

```python
Batch(50000)
```
